```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Computer Vision Masterclass/Images/people1.jpg') 
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
facedetector = cv2.CascadeClassifier('../Computer Vision Masterclass/Cascades/haarcascade_frontalface_default.xml')
detections = facedetector.detectMultiScale(img, scaleFactor=1.2)
print(f"{len(detections)} detections")
for (x,y,w,h) in detections:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 10)
plt.imshow(img)
```
